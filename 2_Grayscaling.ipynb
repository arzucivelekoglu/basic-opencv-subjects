# **Grayscaling**

Grayscaling (Gri Tonlama) renkli görüntüleri (3 kanallı) siyah-beyaz tek kanallı görüntülere dönüştürme işlemidir.

Bu, görüntüyü basitleştirdiği, görüntüde daha az bilgi olduğu için neredeyse bir parazit azaltma ve işleme süresini artırdığı için yapılır.

OpenCV'yi google colaboraty üzerinden çalıştırmak istedğimizde cv.imshow() komutunu kullanamıyoruz. Bu yüzden bu notebookta matplotlib kütüphanesini kullanarak görüntüleri görüntüleyeceğiz.

```python
import cv2
import matplotlib.pyplot as plt

#❗ OpenCV görüntüyü okurken BGR biçiminde okuyor
image = cv2.imread('/content/dog.jpeg')

plt.imshow(image)
```

Gri tonlamaya dönüştürmek için `cvtColor` kullanıyoruz.

 
*   2 bağımsız değişken alır, ilki giriş 
görüntüsüdür
*   İkincisi, renk alanı dönüştürme kodudur.



```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap="gray")
```

## **Grayscale Image Dimensions**

RGB renkli görüntülerin, her ana renk için bir tane olmak üzere 3 boyutu olduğunu unutmayın. Gri tonlama sadece 1'e sahiptir, bu da grinin yoğunluğudur.

image.png

```python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(f"RGB Image Dimensions : {rgb_img.shape}")
print(f"Gray Scale Image Dimensions : {gray_img.shape}")
```

```python
fig = plt.figure(figsize=(10, 7))

fig.add_subplot(1, 2, 1) 
plt.imshow(rgb_img)
plt.axis('off')
plt.title(f"RGB Image\n shape : {rgb_img.shape}")

fig.add_subplot(1, 2, 2)
plt.imshow(gray_img, cmap="gray")
plt.axis('off')
plt.title(f"Gray Image\n shape : {gray_img.shape}")
```
