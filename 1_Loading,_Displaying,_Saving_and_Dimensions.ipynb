### **Loading Images**

`imread` fonksiyonu içerisine görüntünün yolunu belirterek görüntüyü içe aktarıyoruz. 

❗ *OpenCV görüntüyü okurken BGR biçiminde okuyor*

```python
image = cv2.imread("/content/images/Origin_of_Species.jpg")
```

### **Displaying Images**

Görüntüyü matplotlib ile görüntülemek için BGR'dan RGB formatına çevirmemiz gerekiyor.

```python
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Displaying Our Firts Image")
plt.axis("off")
plt.show()
```

### **Saving Images**

`imwrite` fonksiyonu içerisine dosya adını ve kaydedilecek görüntüyü belirtiyoruz. 

```python
cv2.imwrite('output.jpg', image)
```

```python
cv2.imwrite('output.png', image)
```

### **Displaying Image Dimensions**

Görüntüler 3 kanaldan oluşmaktadı: 

image.png


Görüntülerin boyutunu ve şeklini öğrenmek için `Numpy` kütüphanesini kullanıyoruz.

```python
import numpy as np
```

```python
# (row, column, ndim)
# (yükseklik , genişlik, kanal sayısı)
print(f'Shape of Image : {image.shape}')
```

```python
# görüntünün kanal sayısı
print(f'Dimension of Image : {image.ndim}')
```

```python
print('Height of Image: {} pixels'.format(int(image.shape[0])))
print('Width of Image: {} pixels'.format(int(image.shape[1])))
print('Depth of Image: {} colors components'.format(int(image.shape[2])))
```
