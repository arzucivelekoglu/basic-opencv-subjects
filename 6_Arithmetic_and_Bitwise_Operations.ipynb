# **Arithmetic and Bitwise Operations**

## **1.Arithmetic Operations**
Görüntü matrisi üzeirne doğrudan farklı matrisler ekleyip çıkarmamızı sağlıyor.

Giriş görüntülerine Toplama, Çıkarma ve Bitsel İşlemler (AND, OR, NOT, XOR) gibi Aritmetik İşlemler uygulanabilir. Bu işlemler, giriş görüntülerinin özelliklerini geliştirmede yardımcı olabilir.  

Görüntü üzerinde parlaklık arttırma veya azaltma gibi etkileri oluyor.


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("/content/bird.webp")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#görüntü boyutunda 100'lerden oluşan bir matris oluşturduk
# 0-> siyah --- 255->beyaz
M = np.ones(img.shape, dtype = "uint8") * 100 


fig = plt.figure(figsize=(20, 20))

fig.add_subplot(1, 2, 1) 
plt.imshow(img)
plt.title("RGB")

fig.add_subplot(1, 2, 2)
plt.imshow(gray, cmap="gray")
plt.title("Gray")
````

**```
cv2.add()
```
fonksiyonunu kullanarak ekleme**

````python
M = np.ones(img.shape, dtype = "uint8") * 100
added1 = cv2.add(img, M)

M = np.ones(gray.shape, dtype = "uint8") * 100
added2 = cv2.add(gray, M)

fig = plt.figure(figsize=(20, 20))

fig.add_subplot(1, 2, 1) 
plt.imshow(added1)
plt.title("RGB Added")

fig.add_subplot(1, 2, 2)
plt.imshow(added2, cmap="gray")
plt.title("Gray Added")
````

**'+' operatörünü kullanarak ekleme**

````python
M = np.ones(img.shape, dtype = "uint8") * 100
added1 = img + M

M = np.ones(gray.shape, dtype = "uint8") * 100
added2 = gray + M

fig = plt.figure(figsize=(20, 20))

fig.add_subplot(1, 2, 1) 
plt.imshow(added1)
plt.title("RGB Add")

fig.add_subplot(1, 2, 2)
plt.imshow(added2, cmap="gray")
plt.title("Gray Add")
````

**'-' operatörünü kullanarak çıkarma**

````python
M = np.ones(img.shape, dtype = "uint8") * 100
added1 = img - M

M = np.ones(gray.shape, dtype = "uint8") * 100
added2 = gray - M

fig = plt.figure(figsize=(20, 20))

fig.add_subplot(1, 2, 1) 
plt.imshow(added1)
plt.title("RGB")

fig.add_subplot(1, 2, 2)
plt.imshow(added2, cmap="gray")
plt.title("Gray")
````

## **Bitwise Operations and Masking**

````python
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
plt.imshow(square, cmap="gray")
plt.show()

ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
plt.imshow(ellipse,  cmap="gray")
plt.show()
````

AND, OR, XOR ve NOT gibi bazı bitsel işlemlerle denemeler yapmak

````python
And = cv2.bitwise_and(square, ellipse)
plt.imshow(And,  cmap="gray")
plt.title("AND")
plt.show()

# Shows where either square or ellipse is 
bitwiseOr = cv2.bitwise_or(square, ellipse)
plt.imshow(bitwiseOr,  cmap="gray")
plt.title("bitwiseOr")
plt.show()

# Shows where either exist by itself
bitwiseXor = cv2.bitwise_xor(square, ellipse)
plt.imshow(bitwiseXor,  cmap="gray")
plt.title("bitwiseXor")
plt.show()

# Shows everything that isn't part of the square
bitwiseNot_sq = cv2.bitwise_not(square)
plt.imshow(bitwiseNot_sq,  cmap="gray")
plt.title("bitwiseNot_sq")
plt.show()
````

0'dan 1'e kadar değişen alfa ve beta değerini değiştirerek, bir görüntüden diğerine kolaylıkla geçiş yapabiliriz. Bu ağırlıkların değeri 0 ile 1 arasında değişir ve daha sonra ihtiyacımıza göre istenen görüntü görünümüne sahip olabiliriz.

````python
img1 = cv2.imread("/content/bird.webp")
img1 = cv2.resize(img1 , (640,426))

img2 = cv2.imread("/content/bird2.webp")
img2 = cv2.resize(img2 , (640,426))

plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
weightedSum = cv2.addWeighted(img1, 0.4, img2, 0.4, 0)

plt.figure(figsize=(10, 7))
plt.imshow(weightedSum)
````
