## **Re-sizing**

Re-sizing (yeniden boyutlandırma) ile `cv.resize()` işlevini kullanarak görüntülerin şekillerini (shape) değiştirmemizi sağlıyor.


```
cv2.resize(image, dsize(output image size), x scale, y scale, interpolation)
```

Burada ya dzise ile tam istediğiniz boyutu verip ona göre yendiden şekillendirilmesini isteyebiliriz ya da x-y skala değerleri verip ona göre şekillenmesini de isteyebiliriz.



#### **Interpolation Method Yöntemlerinin Listesi:**

- **INTER_NEAREST** –en yakın komşu enterpolasyonu

- **INTER_LINEAR** – çift ​​doğrusal enterpolasyon (varsayılan olarak kullanılır)

- **INTER_AREA** – piksel alanı ilişkisini kullanarak yeniden örnekleme. Haresiz sonuçlar verdiği için görüntü kırpımı için tercih edilen bir yöntem olabilir. Ancak görüntü yakınlaştırıldığında, INTER_NEAREST yöntemine benzer.

- **INTER_CUBIC** – 4 × 4 piksel komşuluk üzerinde çift kübik enterpolasyon

- **INTER_LANCZOS4** – 8 × 8 piksel komşuluk üzerinde bir Lanczos enterpolasyonu


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/default-1.png")

plt.imshow(image)
plt.title(f"Original Image {image.shape[0:2]}")
plt.show()
````

````python
x1 = cv2.resize(image , (15,15), interpolation = cv2.INTER_NEAREST )
x2 = cv2.resize(image , (15,15), interpolation = cv2.INTER_LINEAR )
x3 = cv2.resize(image , (15,15), interpolation = cv2.INTER_AREA )
x4 = cv2.resize(image , (15,15), interpolation = cv2.INTER_CUBIC )
x5 = cv2.resize(image , (15,15), interpolation = cv2.INTER_LANCZOS4 )
````

````python
plt.imshow(image)
plt.title("Original Image")
plt.show()

fig = plt.figure(figsize=(20, 20))

fig.add_subplot(1, 5, 1) 
plt.imshow(x1, cmap="gray")
plt.axis('off')
plt.title(f"INTER_NEAREST\n {x1.shape[0:2]}")

fig.add_subplot(1, 5, 2)
plt.imshow(x2, cmap="gray")
plt.axis('off')
plt.title(f"INTER_LINEAR\n {x2.shape[0:2]}")

fig.add_subplot(1,5, 3)
plt.imshow(x3, cmap="gray")
plt.axis('off')
plt.title(f"INTER_AREA\n {x3.shape[0:2]}")

fig.add_subplot(1,5, 4)
plt.imshow(x4, cmap="gray")
plt.axis('off')
plt.title(f"INTER_CUBIC\n {x4.shape[0:2]}")

fig.add_subplot(1,5, 5)
plt.imshow(x5, cmap="gray")
plt.axis('off')
plt.title(f"INTER_LANCZOS4\n {x5.shape[0:2]}")
````

````python
x = cv2.resize(image ,None, fx= 2, fy=2 , interpolation = cv2.INTER_NEAREST )
plt.imshow(x)
plt.title(f"Shape {x.shape[0:2]}")
````

## **Image Pyraminds**

````python
image = cv2.imread('/content/dog.jpeg')

smaller = cv2.pyrDown(image)
larger = cv2.pyrUp(image)
````

````python
fig = plt.figure(figsize=(15,10))

fig.add_subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title(f"Original {image.shape[0:2]}")

fig.add_subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(smaller, cv2.COLOR_BGR2RGB))
plt.title(f'Smaller {smaller.shape[0:2]}')

fig.add_subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(larger, cv2.COLOR_BGR2RGB))
plt.title(f'Larger {larger.shape[0:2]}')


````

## Cropping

`image[start_row:end_row , start_col:end_col]`

````python
image = cv2.imread("/content/dog.jpeg")

cropped = image[0:300 , 100:350]
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
````

**rectangle**

`cv2.rectangle(image, (start_col,start_row), (end_col,end_row), (0,255,255), 10)`


````python
copy = image.copy()
cv2.rectangle(copy, (100,0),(350,300), (0,0,255),8)
plt.imshow(cv2.cvtColor(copy, cv2.COLOR_BGR2RGB))
````
